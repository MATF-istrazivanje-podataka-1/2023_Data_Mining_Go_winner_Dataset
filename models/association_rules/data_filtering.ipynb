```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../../data/csgo_round_snapshots.csv')
```

```python
data.info()
```

```python
data_filtered = data.drop(columns=['time_left', 'ct_score', 't_score', 'map', 'bomb_planted', 'ct_health', 't_health', 'ct_money', 't_money', 'ct_players_alive', 't_players_alive', 'round_winner'])
```

```python
data_filtered.info()
```

```python
ct_side = [col for col in data_filtered.columns if (col.startswith('ct_'))]
t_side = [col for col in data_filtered.columns if (col.startswith('t_'))]
```

```python
data_ct = data_filtered.drop(columns=t_side)
data_t = data_filtered.drop(columns=ct_side)
```

```python
data_t.head()
```

```python
data_ct.head()
```

```python
data_ct = data_ct.applymap(lambda x: True if x > 0 else False)
data_t = data_t.applymap(lambda x: True if x > 0 else False)
```

```python
data_ct.drop(columns=['ct_weapon_usps', 'ct_weapon_p2000'], inplace=True)
data_t.drop(columns=['t_weapon_glock'], inplace=True)
```
