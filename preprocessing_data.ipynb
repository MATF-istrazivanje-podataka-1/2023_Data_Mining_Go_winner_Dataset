```python
data.drop(index=data.loc[data['t_health'] > 500].index, inplace=True)

for col in data.columns:
    if ('weapon' in col or 'grenade' in col) and data[col].max() > 5:
        print(col, data[col].max())
        data.drop(index=data.loc[data[col] > 5].index, inplace=True)
```

### Podela na ulazne i ciljne podatke
Na kraju svake runde u igrici CS:GO moguca su samo dva ishoda, rundu je pobedila CT strana ili rundu je pobedila T strana. Nase ciljne podatke ce predstavljati upravo krajnji ishod svake runde, koji se nalazi u koloni *round_winner*. Ostatak podataka koristimo kao ulazne podatke kako bi mogli da predvidjamo ko je pobedio rundu.

```python
y = data['round_winner']
X = data.drop('round_winner', axis=1)
feature_names = X.columns
X.head()
```

```python
X.info()
```

```python
y.unique()
```

```python
y.replace('CT', 0, inplace=True)
y.replace('T', 1, inplace=True)
```

```python
y.unique()
```

```python
maps = X['map'].unique()
```

```python
maps_dict = {}
i = 0

for map in maps:
    maps_dict[map] = i
    i += 1

X['map'].replace(maps_dict, inplace=True)
X
```

```python
X['bomb_planted'].replace({True: 1, False: 0}, inplace=True)
X
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=55)
```

```python
X_train
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)

X_train = pd.DataFrame(data=X_train_scaled, columns=X_train.columns)
X_test = pd.DataFrame(data=X_test_scaled, columns=X_test.columns)
```

```python
X_train.describe()
```

```python
from joblib import dump
```

```python
dump((X_train, X_test, y_train, y_test), 'data.joblib')
```
